Load AirBnB dataset into Pandas Data Frame:

```python
df=pd.read_csv('C:/Users/Sidney/OneDrive/Documents/LiveProject/AB_NYC_2019.csv')
```

List names of columns of dataframe:

```python
Headings=list(df)
#list(df) is a nice list of all the column headings
```

```python
Headings
```

```python
is_nulldf=df.isnull()
#how do you get a list of each column for the dataframe df.isnull()?
```

```python
is_nulldf
```

```python
for i in range(len(Headings)):
    print()
    print(Headings[i],':')
    z=list(is_nulldf[Headings[i]])
    y=list(df[Headings[i]])
    count=z.count(True)
    print(f'null values in column {Headings[i]} is {count}')
    
    if Headings[i] in continuous:
        neg_count=0
        for j in range(len(z)):
            if (z[j]==False) and (y[j]<0):
                neg_count += 1
        print(f'neg values in column {Headings[i]} is {neg_count}') 
        
    if Headings[i]=='latitude':
        count_outside=0
        for j in range(len(z)):
            if (z[j]==False) and ((y[j] < min_lat) or (y[j] > max_lat)):
                count_outside += 1
        print(f'values outside NYC latitude are {count_outside}')
        
    if Headings[i]=='longitude':
        count_outside=0
        for j in range(len(z)):
            if (z[j]==False) and ((y[j] < min_long) or (y[j] > max_long)):
                count_outside += 1
        print(f'values outside NYC longitude are {count_outside}')
    
    if Headings[i] in cat:
        categories=list(df[Headings[i]].value_counts())
        print(f'Categories in column {Headings[i]} are {len(categories)}')
   
```
